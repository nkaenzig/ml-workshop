#### Processing  & Transformations

```python
# | sampling
df = df.sample(frac=0.2)
```

```python
# | shuffling
df = df.sample(frac=1)
```

```python
# | eliminar filas con valores faltantes
df = df.dropna()
```

```python
# | eliminar columas con valores faltantes
df = df.dropna(axis=1)
```

```python
# | eliminar filas duplicadas
df = df.drop_duplicates()
```

```python
# | sorting
df = df.sort_values(by='age', ascending=False)
df
```

## Matplotlib & Seaborn

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(100, 4), columns = list("ABCD"))
df.head()
```

### Integración de matplotlib en Pandas

```python
import pandas as pd
```

```python
df.plot()
```

```python
df['A'].plot.hist()
```

```python
df.plot.scatter(x='A', y='B')
```

```python
df.plot.box()
```

### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.plot(df)
```

```python
fig = plt.hist(df['B'])
```

```python
fig = plt.scatter(x=df['A'], y=df['B'])
```

```python
fig = plt.boxplot(df.values)
```

#### Subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(7,3))
plt.tight_layout()

_ = axs[0][1].hist(df['A'])
_ = axs[1][0].plot(df)
_ = fig.axes[3].boxplot(df.values)
```

### Seaborn

```python
import seaborn as sns
```

```python
# | Para visualizar las correlaciones y distribuciones de todas las columnas
sns.pairplot(df.sample(100))
```

```python
# | Para visualizar las (pearson) correlaciones de todas las columnas
sns.heatmap(df.corr().abs(), annot=True, cmap="Blues")
```
